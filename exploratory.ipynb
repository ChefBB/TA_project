```python
import pandas as pd

path = '/Users/brunobarbieri/Library/CloudStorage/OneDrive-UniversityofPisa/TA_Project/data/'
df = pd.read_csv(path + 'lab_lem_merged.csv')
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize= (10,6))
df.groupby('label').count()['id'].sort_values(ascending= False).plot(kind= 'bar')
plt.xlabel('Class')
plt.ylabel('Number of rows')
plt.xticks(rotation= 0.9)
plt.title('Number of occurrences for each class')
plt.grid(axis= 'y', alpha= 0.3)
```
